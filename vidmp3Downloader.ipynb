```python

def download(Link):
    cwdpath = os.getcwd()
    print(f' Your location : {cwdpath}')

    # DOWNLOAD YOUTUBE VIDEOS
    print(' Let\'s download a Video ')
    yt = YouTube(Link)
    title = yt.title
    print(f' Video Name      : {title}')
    desc = yt.description
    lenght = yt.length
    print(f' Length of video : {lenght} seconds')

    video = yt.streams.get_highest_resolution()
    vid_loc = os.path.join(cwdpath,'contents',title.replace(' ',''))
    path = video.download(vid_loc)

    document = {
        'Title'          : title,
        'Description'    : desc,
        'Video_Path'     : path,
        'Video_Folder' : vid_loc
    }
    return document


def split(meta):
  #SPLIT VIDEO
  path  = meta['Video_Path'] 
  print('\n ---------------------------------------------\n')
#check for py  for videosplitter
  splitter = '/content/video-splitter/ffmpeg-split.py'
  if os.path.isfile(splitter) :
    print('\n File is available. Moving along.')
  else:
    !git clone https://github.com/c0decracker/video-splitter.git
  print('\n ---------------------------------------------\n')

  subprocess.run(['python', splitter ,'-f', path,'-s','59' ])
  os. remove(path) 

def file_retrieve(meta):
  Path  = meta['Video_Folder'] 
  lists = os.listdir(Path)
  x= 0
  paths = []
  for i in lists:
    video_path = os.path.join(Path,i)
    reg = int(re.findall('(-\d-\d*)', i)[0].replace('-',''))
    path_Dict = {
        'Path'      : video_path,
        'Part'      : reg,
        'Vid_title' : meta['Title'],
        'Total_pts' : len(lists)
    }
    print(f' {i} part : {reg}')
    paths.append(path_Dict)
    x+=1
  return paths

  
def MP3_PDownloader(P_Link):
  playlist = P_Link#Playlist(P_Link)
  #prints each video url, which is the same as iterating through playlist.video_urls
  cwd = os.getcwd()
  mp3_folder = os.path.join(cwd,'customSongs')
  video_folder = os.path.join(cwd,'customSongs','VIDEOS')
  try:
    os.mkdir(mp3_folder)
    os.mkdir(video_folder)
  except:
    print('Folder : customSongs Found.')
  #  path =  r'/content/customSongs'
  #For saving mp3 files.

  try :
    os.mkdir(os.path.join(mp3_folder,'SONGS'))
  except:
    print(' Folder : Songs Exists already. ')

 
  for url in playlist:
      print('Playlist : ',url)
      yt = YouTube(url)

      print('---------------------------------------------')
      try : 
        print(f' Downloading : {yt.title}')
      except:
        continue
      downloadpath = yt.streams.first().download(video_folder)
      print('---------------------------------------------')


      # print(f' Download path : {downloadpath}')
      
      SPLIT_PATH = downloadpath.split('/')
      P = len(SPLIT_PATH)-1
      # print(f' Path = {SPLIT_PATH}')

      

      print(' - Converting mp4 to mp3 - ')
      # /content/customSongs/The Chainsmokers - Dont Let Me Down (Illenium Remix).mp4
      
      try :
        name, ext = SPLIT_PATH[P].split('.')
        if (ext =='mp4'):
          print(f' {name} is converting to mp3 now.')
          #  print('\n---------------------------------------------')
          # print('\n#############################################')
          subprocess.run(['ffmpeg','-i', downloadpath, os.path.join(mp3_folder,'SONGS', f'{name}.mp3')])
          # print(' Disposing of mp4 file. ')
          # os.remove(downloadpath)

      except:
          print('Something wrong... ')
      print('\n---------------------------------------------')
      time.sleep(2.4)
  print(' All work is done')
  
  #Important Zipping musics to download
  print(' Please Wait while I zip for download. \n\n')
  !zip -r /content/MP3AUDIO.zip /content/customSongs/SONGS
  print('$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$')
  print('\tR E A D Y  F O R  D O W N L O A D !')


# def main():
#     Link = input(' Please Enter YouTube Link : ')
#     meta = download(Link)
#     #https://www.youtube.com/watch?v=d5ZFBbmvlSQ
#     split(meta)
#     file_retrieve(meta)
```
